### Question1:

a) 

$$ f_{1}(x)=a_{1} +b_{1}x+c_{1}x^{2} +d_{1}x^{3}$$
$$For\space x<\zeta:$$
$$f(x)=β_{0} +β_{1}x+β_{2}x^{2} +β_{3}x^{3}$$
$$ Then, $$
$$ a_{1}=β_{0},b_{1}=β_{1},c_{1}=β_{2},d_{1}=β_{3}$$

b)

$$ f_{2}(x)=a_{2} +b_{2}x+c_{2}x^{2} +d_{2}x^{3}$$
$$For\space x>\zeta:$$
$$f(x)=β_{0} +β_{1}x+β_{2}x^{2} +β_{3}x^{3}+β_{4}(x-\zeta)^{3}$$
$$=β_{0} +β_{1}x+β_{2}x^{2} +β_{3}x^{3}+β_{4}(x^{3}-3x^{2}\zeta+3x\zeta_{2}-\zeta^{3})$$
$$=(β_{0}-β_{4}\zeta^{3})+(β_{1}+3β_{4}\zeta^{2})x+(β_{2}-3β_{4}\zeta)x^{2}+(β_{3}+β_{4})x^{3}$$
$$ Then, $$
$$ a_{2}=β_{0}-β_{4}\zeta^{3} \space,\space b_{2}=β_{1}+3β_{4}\zeta^{2} \space,\space c_{2}=β_{2}-3β_{4}\zeta \space,\space d_{2}=β_{3}+β_{4}$$

c)

$$When \space x=\zeta$$
$$ f_{1}(x)=β_{0} +β_{1}\zeta+β_{2}\zeta^{2} +β_{3}\zeta^{3}$$
$$ f_{2}(x)=(β_{0}-β_{4}\zeta^{3}) +(β_{1}+3β_{4}\zeta^{2})\zeta+(β_{2}-3β_{4}\zeta) \zeta^{2} +(β_{3}+β_{4})\zeta^{3}$$
$$ =β_{0} +β_{1}\zeta+β_{2}\zeta^{2} +β_{3}\zeta^{3} $$
$$ then \space f_{1}(x)=f_{2}(x) $$

d)

$$When \space x=\zeta$$
$$ f^{'}_{1}(x)=b_{1}+2c_{1}x +3d_{1}x^{2}$$
$$ f^{'}_{1}(x)=β_{1}+2β_{2}\zeta +3β_{3}\zeta^{2}$$

$$ f^{'}_{2}(x)=b_{2}+2c_{2}x +3d_{2}x^{2}$$
$$ =β_{1}+3β_{4}\zeta^{2}+2(β_{2}-3β_{4}\zeta) \zeta +3(β_{3}+β_{4})\zeta^{2} $$
$$=β_{1}+2β_{2}\zeta +3β_{3}\zeta^{2}$$
$$ then \space f^{'}_{1}(x)=f^{'}_{2}(x) $$

e)

$$When \space x=\zeta$$
$$ f^{''}_{1}(x)=2c_{1} +6d_{1}x$$
$$ f^{''}_{1}(x)=2β_{2} +6β_{3}\zeta$$

$$ f^{''}_{2}(x)=2(β_{2}-3β_{4}\zeta) +6(β_{3}+β_{4})\zeta$$
$$=2β_{2} +6β_{3}\zeta$$
$$ then \space f^{''}_{1}(x)=f^{''}_{2}(x) $$

There will be no penalty term so function g will be very jumpy and will exactly interpolate the training observations.

$$ When\space \lambda->\infty $$
The smoothing spline $ĝ_{2}$ will probably have the smaller training RSS because it will be a higher order polynomial due to the order of the penalty term (it will be more flexible).

b)
$$ When\space \lambda->\infty $$

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
X_df = wage_df['age']
y_df = wage_df['wage']
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2)
```

```python
def Polynomial_CV(X_train, Y_train, X_test, Y_test, M, splits):
    
    val_MSE = {}
    test_MSE = {}

    for m in M: 
        mse = 0
        kf = KFold(n_splits=splits)
        poly = PolynomialFeatures(degree=m)
        X_train_p = poly.fit_transform(np.array(X_train).reshape(-1, 1))
        for train_index, test_index in kf.split(X_train_p):
            X, X_CV = X_train_p[train_index], X_train_p[test_index]
            Y, Y_CV = Y_train.iloc[train_index], Y_train.iloc[test_index]
            model = LinearRegression(fit_intercept=True)
            model.fit(X, Y)
            p = model.predict(X_CV)
            mse += mean_squared_error(p, Y_CV)
        val_MSE[m] = mse/splits

        # Compute test MSE for the model
        model = LinearRegression(fit_intercept=True)
        model.fit(X_train_p, Y_train)
        p = model.predict(poly.fit_transform(np.array(X_test).reshape(-1, 1)))
        test_MSE[m] = mean_squared_error(p, Y_test)


    min_val_components = min(val_MSE, key=val_MSE.get)
    min_test_components = min(test_MSE, key=test_MSE.get)
    # Plot validation MSE
    lists = sorted(val_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(121)
    plt.plot(x, y, color='r')
    plt.axvline(x=min_val_components,ls='--',color='b')
    plt.grid()
    ax.set_xlabel('Degree')
    ax.set_ylabel('Validation MSE')
    ax.set_title('Validation MSE vs Degree')
    

    lists = sorted(test_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    ax = fig.add_subplot(122)
    plt.plot(x, y, color='g')
    plt.axvline(x=min_val_components,ls='--',color='r')
    plt.axvline(x=min_test_components,ls='--',color='b')
    
    plt.grid()
    ax.set_xlabel('Degree')
    ax.set_ylabel('Test MSE')
    ax.set_title('Test MSE vs Degree')
    plt.show()
    
    return val_MSE,test_MSE
```

```python
val_MSE,test_MSE = Polynomial_CV(X_train,y_train,X_test,y_test,np.arange(1,14,1),10)
```

```python
import statsmodels.api as sm
def anova_models(X, Y, M):
    models={}
    for m in M:
        poly = PolynomialFeatures(degree=m)
        X_train_p = poly.fit_transform(np.array(X).reshape(-1, 1))
        model = sm.OLS(Y, sm.add_constant(X_train_p))
        model = model.fit()
        models[m] = model
    table = sm.stats.anova_lm(models[1],models[2],models[3],models[4],models[5],models[6],models[7])
    return table
```

```python
def step_CV(X_train, Y_train, X_test, Y_test, M, splits):
    val_MSE = {}
    test_MSE = {}

    for i in M: 
        mse = 0
        kf = KFold(n_splits=splits)
        age_groups = pd.cut(X_train, i)
        df_dummies = pd.get_dummies(age_groups)
        
        for train_index, test_index in kf.split(df_dummies):
            X, X_CV = df_dummies.iloc[train_index], df_dummies.iloc[test_index]
            Y, Y_CV = Y_train.iloc[train_index], Y_train.iloc[test_index]
            model = sm.GLM(Y, sm.add_constant(X)).fit()
            
            p = model.predict(sm.add_constant(X_CV))
            mse += mean_squared_error(p, Y_CV)
        val_MSE[i] = mse/splits

        # Compute test MSE for the model
        model = sm.GLM(Y_train, sm.add_constant(df_dummies)).fit()
        test_age = pd.cut(X_test, i)
        test_df_dummies = pd.get_dummies(test_age)
        p = model.predict(sm.add_constant(test_df_dummies))
        test_MSE[i] = mean_squared_error(p, Y_test)


    min_val_components = min(val_MSE, key=val_MSE.get)
    min_test_components = min(test_MSE, key=test_MSE.get)
    # Plot validation MSE
    lists = sorted(val_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(121)
    plt.plot(x, y, color='r')
    plt.axvline(x=min_val_components,ls='--',color='b')
    plt.grid()
    ax.set_xlabel('Cuts')
    ax.set_ylabel('Validation MSE')
    ax.set_title('Validation MSE vs Cuts')
    

    lists = sorted(test_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    ax = fig.add_subplot(122)
    plt.plot(x, y, color='g')
    plt.axvline(x=min_val_components,ls='--',color='r')
    plt.axvline(x=min_test_components,ls='--',color='b')
    
    plt.grid()
    ax.set_xlabel('Cuts')
    ax.set_ylabel('Test MSE')
    ax.set_title('Test MSE vs Cuts')
    plt.show()
    
    return val_MSE,test_MSE
```

```python
val_MSE,test_MSE = step_CV(X_train,y_train,X_test,y_test,np.arange(1,20,1),10)
```

```python
def add_polt(ax, x_true, x_pred, y_true, y_pred, RSS, degree):
    ax.plot(x_pred, y_pred, 'red', label="Degree: " +str(degree))
    ax.scatter(x_true, y_true, alpha=1, color='green')
    
    ax.set_xlabel('dis')
    ax.set_ylabel('nox')
    ax.set_title('Degree: ' +str(degree) + ' Residuals: ' + str(RSS))
    ax.legend(loc='best')
    ax.grid()
    return None
def multi_models_with_plots(X,y,M):
    fig = plt.figure(figsize=(30, 40))
    x = np.linspace(X.min()-1, X.max()+1, 256, endpoint = True)
    RSS={}
    for m in M:
        degree=m
        poly = PolynomialFeatures(degree=m)
        X_ = poly.fit_transform(np.array(X).reshape(-1,1))
        x_pred= poly.fit_transform(np.array(x).reshape(-1,1))
        
        model = sm.OLS(y, X_)
        result = model.fit()
        params = np.array(result.params)
        y_pred = np.dot(x_pred,params).reshape(-1,1)
        RSS[m] = result.ssr
        ax = fig.add_subplot(6, 2, degree)
        
        add_polt(ax, X, x, y, y_pred, result.ssr, degree)
   
    
    lists=sorted(RSS.items())
    x,y = zip(*lists)
    ax = fig.add_subplot(6, 2, M[-1]+1)
    ax.plot(x,y,color='r')
    
    plt.show()
        
    
        
        
    
```

```python
multi_models_with_plots(boston_df['dis'],boston_df[['nox']],np.arange(1,11,1))
```

```python
def splines_with_df(X,y,D):
    RSS = {}
    for d in D:
        transformed_x2 = dmatrix("bs(X, df=d, include_intercept=False)",
                        {"X": X}, return_type='dataframe')
        glm_fit = sm.GLM(y, transformed_x2).fit()
        
        y_pred = glm_fit.predict(transformed_x2)
        RSS[d] = mean_squared_error(y,y_pred)*len(y)
        
    min_val_components = min(RSS, key=RSS.get)
   
    lists=sorted(RSS.items())
    x,y = zip(*lists)
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(121)
    plt.plot(x, y, color='r')
    plt.axvline(x=min_val_components,ls='--',color='b')
    plt.grid()
    ax.set_xlabel('Spline_degree')
    ax.set_ylabel('RSS')
    ax.set_title('RSS vs Spline_degree')
    
    
    plt.show()
        


    
```

```python
splines_with_df(boston_df['dis'], boston_df['nox'],np.arange(3,17,1))
```

###### We may see that RSS decreases until 14 and then slightly increases after that so the best number of degrees of freedom is 14.

```python
def splines_cv(X,Y,D,splits):
    RSS = {}
    MSE = {}
    for d in D:
        mse = 0
        kf = KFold(n_splits=splits)
        for train_index, test_index in kf.split(X):
            x, x_CV = X[train_index], X[test_index]
            y, y_CV = Y.iloc[train_index], Y.iloc[test_index]
            transformed_x = dmatrix("bs(x, df=d, include_intercept=False)",
                        {"x": x}, return_type='dataframe')
            glm_fit = sm.GLM(y, transformed_x).fit()
            transformed_x_cv = dmatrix("bs(x_CV, df=d, include_intercept=False)",
                        {"Xx_CV": x_CV}, return_type='dataframe')
            y_pred = glm_fit.predict(transformed_x_cv)
            mse+= mean_squared_error(y_CV,y_pred)
        MSE[d] = mse/splits
        RSS[d] = MSE[d]*len(y)
        
    min_val_components = min(RSS, key=RSS.get)
   
    lists=sorted(RSS.items())
    x,y = zip(*lists)
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(121)
    plt.plot(x, y, color='r')
    plt.axvline(x=min_val_components,ls='--',color='b')
    plt.grid()
    ax.set_xlabel('Spline_degree')
    ax.set_ylabel('RSS')
    ax.set_title('RSS vs Spline_degree')
    
    
    plt.show()
```

```python
splines_cv(boston_df['dis'], boston_df['nox'],np.arange(3,17,1),10)
```

```python
def fit_linear_reg_sm(X,Y):
    #Fit linear regression model and return RSS and adjusted R squared values, AIC, BIC, MAllow's Cp
    model = sm.OLS(Y,sm.add_constant(X)).fit()
    n=len(Y)
    num_params=X.shape[1]+1
    RSS = mean_squared_error(Y,model.predict(sm.add_constant(X))) * n
    adj_R_squared = model.rsquared_adj
    aic = model.aic
    bic = model.bic
    cp = (RSS+2*num_params*(RSS/n))/n
    return [RSS, adj_R_squared, aic, bic, cp,model]
```

```python
def forward_stepwise(X, y,scorer='rss',threshold=0.01,num_features=3):
    p_used = []
    RSS_list,R_squared_list,Aic_list,Bic_list,cp_list,feature_list=[],[],[],[],[],[]
    numb_features = []
    models_list=[]
    
    
    for k in range(1,len(X.columns)+1):
        best_RSS=np.inf
        best_adj_r_squared=0
        best_cp=np.inf
        best_bic=np.inf
        best_aic=np.inf
        p_left = list(set(X.columns)-set(p_used))
        #print(p_left)
        for combo in itertools.combinations(p_left,1):
            #print(combo)
            x=X[p_used+list(combo)]
            tmp_result = fit_linear_reg_sm(x,y)
            #print(tmp_result[1])
            if scorer=='rss':
                if (tmp_result[0]< best_RSS):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
            
            elif scorer=='adj_r':
                
                if (tmp_result[1]> best_adj_r_squared): #and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_adj_r_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
                
                    
            elif scorer=='aic':
                if (tmp_result[2]< best_aic):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
                    
            elif scorer=='bic':
                if (tmp_result[3]< best_bic):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
                    
            elif scorer=='cp':
                if (tmp_result[4]< best_cp):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
        #print(best_feature)
        if (best_feature not in p_used) and (len(p_used)<num_features) :
            p_used.append(best_feature)
            p_left.remove(best_feature)
            RSS_list.append(best_RSS)
            R_squared_list.append(best_R_squared)
            Aic_list.append(best_aic)
            Bic_list.append(best_bic)
            cp_list.append(best_cp)
            models_list.append(best_model)
            feature_list.append(p_used.copy())
            numb_features.append(len(p_used))
        
        
    df = pd.DataFrame({'numb_features': numb_features,'RSS': RSS_list, 
                   'Adj_R_squared':R_squared_list,'AIC':Aic_list,'BIC':Bic_list,
                   'Cp':cp_list,'features':feature_list,'Model':models_list})
            

        
    return df
```

```python
def backfitting(y,x_s,betas,n_iterations=1000,epsilon=1e-7):
    betas_change={}
    i=0
    Y= y.reshape(-1,1)
    betas_change[i]=betas.copy()
    while i <= n_iterations:
        #print(i)
        
        for j in range(len(betas)):#beta
            #print(j)
            beta_old = betas[j][0]
            X_ = x_s[:,j]
            #print(X_)
            X = np.concatenate((np.array(x_s[:,:j]),np.array(x_s[:,j+1:])),axis=1)
            #print(X)
            Beta = np.concatenate((np.array(betas[:j,:]),np.array(betas[j+1:,:])),axis=0)
            #print(Beta)
            #print(np.dot(X,Beta))
            a = Y-np.dot(X,Beta)
            #print(a)
            model = sm.OLS(a,X_).fit()
            #print(model.summary())
            beta_new = model.params[0]
            #print(beta_old)
            #print(beta_new)
            if np.abs(beta_new-beta_old) < epsilon:
                continue
            else:
                betas[j][0] = beta_new
        
        if i>0:
            
            betas_change[i]=betas.copy()
            #print(betas_change)
            #print(betas_change[i-1])
            y = np.abs(np.subtract(betas_change[i], betas_change[i-1],dtype=np.float))
            #print(y.sum())
            if (np.sum(y)<len(betas)*epsilon):
                print(f"Convergence happened at {i} iteration")
                break
        i+=1   
            
    return betas,betas_change
            
                
            
            
        
    
    
```
