```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import ProbPlot
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

%matplotlib inline

weekly_df=pd.read_csv("csv/Weekly.csv")
weekly_df.tail()
```

```python
weekly_df.describe()
```

```python
#A)
sns.pairplot(data=weekly_df)
```

```python
plt.figure(figsize=[10,8])
sns.heatmap(weekly_df.corr(),annot=True,cmap='viridis_r');
```

###### The correlations between the “lag” variables and today’s returns are close to zero. The only substantial correlation is between “Year” and “Volume”. When we plot “Volume”, we see that it is increasing over time.

```python
#B)
weekly_df['Direction']=weekly_df['Direction'].map({'Down':0,'Up':1})
model_1 = smf.glm(formula='Direction ~ Lag1+Lag2+Lag3+Lag4+Lag5+Volume', data=weekly_df,family=sm.families.Binomial()).fit()
print(model_1.summary())
```

##### It seems that Lag2 is the only predictor taht is statistically significant.

Side note for myself:If we use The direction without mapping the values to 0 and 1 before fitting the model the model will take the vvalues by order assigning 1 to Down because the first value met in our data is Down and o for up so all the coefficients will be the sane but with negative sign

```python
print(model_1.params)
```

```python
X=sm.add_constant(weekly_df.drop(['Direction','Year','Today'],axis=1))

probs=(model_1.predict(X))
predictions=probs>0.5
```

```python
#C)
conf_mat=confusion_matrix(weekly_df['Direction'], predictions)
conf_mat
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.matshow(conf_mat, cmap=plt.cm.Reds, alpha=0.7)
for i in range(conf_mat.shape[0]):
    for j in range(conf_mat.shape[1]):
        ax.text(x=j, y=i,
        s=conf_mat[i, j],
        va='center', ha='center')
plt.xlabel('predicted label')
plt.ylabel('true label')
plt.title("Confusion Matrix")
plt.show()
```

```python
#Another way to display confusion matrix using sklearn library
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat,display_labels=['Down','Up'])
disp.plot(values_format="2g");
```

###### We may conclude that the percentage of correct predictions on the training data is (54+557)/1089 wich is equal to 56.1065197%. In other words 43.8934803% is the training error rate.
###### We could also say that for weeks when the market goes up, the model is right 92% of the time. For weeks when the market goes down, the model is right only 11% of the time, so we can be confident when our model tells us that the market is going up but we can't be this sure when it goes down.

```python
auto_df=pd.read_csv("./csv/Auto.csv")
auto_df.head()
```

```python
#A)
mpg_median=auto_df['mpg'].median()
auto_df['mpg_01']=auto_df['mpg'].apply(lambda x:1 if x>mpg_median else 0)
```

```python
#B)
sns.pairplot(data=auto_df)
```

```python
plt.figure(figsize=[10,8])
sns.heatmap(auto_df.corr(),annot=True);
```

```python
#C)
from sklearn.model_selection import train_test_split
X=auto_df[['cylinders','displacement','horsepower','weight','year']]
Y=auto_df['mpg_01']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
#D)
LDA_model=LinearDiscriminantAnalysis()
model_1=LDA_model.fit(X_train,y_train)
model_1.coef_
```

```python
def get_conf(y_test,y_pred,labels_output):
    t_p=y_test[(y_pred==y_test) & (y_test==1)].sum()
    t_n=y_test[(y_pred==y_test) & (y_test==0)].sum()
    f_p=y_test[(y_pred!=y_test) & (y_test==0)].sum()
    f_n=y_test[(y_pred!=y_test) & (y_test==1)].sum()
    print("Accuarcy of the model="+str((t_p+t_n)/(t_p+t_n+f_p+f_n)))
    print("Error of the model="+str((f_p+f_n)/(t_p+t_n+f_p+f_n)))
    conf_mat=confusion_matrix(y_test,y_pred)
    print(conf_mat)
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat,display_labels=labels_output)
    disp.plot(values_format="2g",cmap='viridis')
    plt.show()
```

```python
y_pred=model_1.predict(X_test)
```

```python
get_conf(y_test,y_pred,[0,1])
```

```python
#E)
QDA_model=QuadraticDiscriminantAnalysis()
model_2=QDA_model.fit(X_train,y_train)
y_pred=model_2.predict(X_test)
```

```python
get_conf(y_test,y_pred,[0,1])
```

```python
#F)
from sklearn.linear_model import LogisticRegression
model_3=LogisticRegression(random_state=0,penalty='l1',solver='liblinear').fit(X_train,y_train)
y_pred=model_3.predict(X_test)
```

```python
get_conf(y_test,y_pred,[0,1])
```

```python
#G)
for k in range(1,23):
    print(k)
    KNN_model=KNeighborsClassifier(n_neighbors=k)
    model1_knn=KNN_model.fit(X_train,y_train)
    y_pred=model1_knn.predict(X_test)
    get_conf(y_test,y_pred,[0,1])
    
```

K=3 seems to be the best value for k taht has the best performance on our data

```python
#D)
def power_3(x,a):
    result=np.power(x,a)
    return result
```

```python
#E)
x=list(range(1,11))
y=power_3(x,3)
ax = sns.scatterplot(x=x, y=np.log(y))
plt.xlabel('x')
plt.ylabel('$log(x^2)$')
plt.title('Relationship of $log(x^2)$ against x')
plt.show();
```

```python
def PlotPower(start,stop,power):
    x=np.arange(start,stop+1)
    y=power_3(x,power)
    ax = sns.scatterplot(x=x, y=y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Relationship of $x^{}$ against x'.format(power))
    plt.show();
```
