```python
from sklearn import svm

X = [[3, 2, 4, 1, 2, 4, 4], [4, 2, 4, 4, 1, 3, 1]]
Y = ['r', 'r', 'r', 'r', 'b', 'b', 'b']
clf = svm.SVC(kernel='linear', C=100)
clf.fit(np.array(X).T.tolist(), Y)

#Decision boundary equation
inter = clf.intercept_[0]

coef_0 = clf.coef_[0][0]

coef_1 = clf.coef_[0][1]
X_1=np.linspace(0,5,100)
X_2=(-inter-coef_0*X_1)/coef_1
print(inter,coef_0,coef_1)
# plot the parallels to the separating hyperplane that pass through the
# support vectors
b = clf.support_vectors_[0]
X2_down = (-coef_0/coef_1) * X_1 + (b[1] - (-coef_0/coef_1) * b[0])
b = clf.support_vectors_[-1]
X2_up = (-coef_0/coef_1) * X_1 + (b[1] - (-coef_0/coef_1) * b[0])
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=180, facecolors='none', edgecolors='black')
plt.scatter(X[0], X[1], c=Y)
plt.plot(X_1, X_2, 'k-')
plt.plot(X_1, X2_down, 'k--', color='g')
plt.plot(X_1, X2_up, 'k--', color='g')

ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_xlabel('X1')
ax.set_ylabel('X2')

plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC,LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
import warnings
warnings.filterwarnings('ignore')
np.random.seed(0)
%matplotlib inline
```

```python
np.random.seed(0)
X = np.random.normal(0, 2, 100)
Y = X**2 + 3*X + np.random.normal(0, 0.2, 100)

c = list(range(0, 100))
c1 = np.random.randint(0, 100, size=50, dtype='l')
c2 = [x for x in c if x not in c1]
Y[c1] = Y[c1] + 5
Y[c2] = Y[c2] - 5

labels = np.asarray([1]*100)
labels[c2] = labels[c2] -2

M = np.column_stack((X,Y))

X_train, X_test, y_train, y_test = train_test_split(M, labels, test_size=0.1)
```

```python
plt.scatter(M[:,0], M[:,1], s=50, c=labels ,cmap="viridis")
plt.xlabel('X1')
plt.ylabel('X2');
```

```python
def make_meshgrid(x, y, h=0.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out



# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
models = [
    SVC(kernel="linear", C=C),
    SVC(kernel="rbf", gamma=0.7, C=C),
    SVC(kernel="poly", degree=2, gamma="auto", C=C),
    SVC(kernel="poly", degree=3, gamma="auto", C=C),
]
models = [clf.fit(X_train, y_train) for clf in models]

# title for the plots
titles = [
    "SVC with linear kernel",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 2) kernel",
    "SVC with polynomial (degree 3) kernel",
]

# Set-up 2x2 grid for plotting.
f = plt.figure(figsize=(10,10))


X0, X1 = M[:, 0], M[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, value in zip(models, titles, [221,222,223,224]):
    ax = f.add_subplot(value)
    plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=labels, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel("X1")
    ax.set_ylabel("X2")
    
    ax.set_title(title)

plt.show()
plt.tight_layout()
```

```python
train_error={titles[i]:1-accuracy_score(y_train,models[i].predict(X_train)) for i in range(len(models))}
```

```python
test_error={titles[i]:1-accuracy_score(y_test,models[i].predict(X_test)) for i in range(len(models))}
```

```python
for i in train_error.keys():
    print(f"Train error rate for {i} : {round(train_error[i]*100,2)}")
    print(f"Test error rate for {i} : {round(test_error[i]*100,2)}")
```

###### This simple example shows a situation in which the linear SVC doesn't work as well as the other kernels (polynomial and radial) on the training data. But on the test set the linear kernel performs well but even better than the SVM with polynomial (degree 2).

```python
np.random.seed(0)
X1 = np.random.uniform(0, 10, 1000) -0.3
X2 = np.random.uniform(0, 10, 1000) -0.3
X = np.column_stack((X1,X2))

temp = X1-X2
Y = [None] * len(temp)
# Assign class labels
for idx, j in enumerate(temp):
    if j > 1:
        Y[idx] = 1
    elif j <  -1:
        Y[idx] = 0
    else:
        Y[idx] = 1 if np.random.uniform(0, 1) > 0.01 else 0
Y = np.array(Y)
color= ['red' if l == 0 else 'blue' for l in Y]

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

plt.scatter(X1, X2, color=color)
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import KFold
cost = [0.0001, 0.001, 0.01, 0.1, 1, 2, 3, 5, 10, 100, 1000, 10000]
cross_vald_error = {}
training_error = {}
test_error= {}

for c in cost:
    accuarcy=0
    kf = KFold(n_splits=10)
    
    
    for train_index, test_index in kf.split(X_train):
        
        X_t, X_CV = X_train[train_index], X_train[test_index]
        
        Y_t, Y_CV = y_train[train_index], y_train[test_index]
        clf = SVC(kernel='linear', C=c)
        clf.fit(X_t, Y_t)
        p = clf.predict(X_CV)
        accuarcy += accuracy_score(p, Y_CV)
    val_acc = accuarcy/10
    cross_vald_error[c]=1-val_acc
    clf.fit(X_train, y_train)
    p_train = clf.predict(X_train)
    training_error[c] = 1- accuracy_score(y_train,p_train) 
    
    p_test = clf.predict(X_test)
    test_error[c] = 1- accuracy_score(y_test,p_test) 

lists = sorted(training_error.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

ax.set_xscale('log')
plt.plot(x, y, color='r', label="Training Error Rate")
ax.set_xlabel("Cost (log)")
ax.set_ylabel("Error Rate")

lists = sorted(cross_vald_error.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

ax.set_xscale('log')
plt.plot(x, y, color='g', label="Validation Error Rate")
ax.set_xlabel("Cost (log)")
ax.set_ylabel("Error Rate")

lists = sorted(test_error.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

ax.set_xscale('log')
plt.plot(x, y, color='b', label="Test Error Rate")
ax.set_xlabel("Cost (log)")
ax.set_ylabel("Error Rate")

plt.grid()
plt.legend()
plt.show()
```

##### AS we can see when C increases the model tries to correctly classify noisy-points and hence overfits the train data. Small C, however, makes a few errors on the noisy test points and performs better on test data.

```python
f = plt.figure(figsize=(15,15))
for degree, value in zip(degrees.keys(), [221,222,223,224]):
    training_error = degrees[degree][0]
    cross_vald_error = degrees[degree][1]
    test_error = degrees[degree][2]
    lists = sorted(training_error.items()) # sorted by key, return a list of tuples

    x, y = zip(*lists) # unpack a list of pairs into two tuples
    ax = f.add_subplot(value)

    ax.set_xscale('log')
    plt.plot(x, y, color='r', label="Training Error Rate")
    ax.set_xlabel("Cost (log)")
    ax.set_ylabel("Error Rate")

    lists = sorted(cross_vald_error.items()) # sorted by key, return a list of tuples

    x, y = zip(*lists) # unpack a list of pairs into two tuples

    ax.set_xscale('log')
    plt.plot(x, y, color='g', label="Validation Error Rate")
    ax.set_xlabel("Cost (log)")
    ax.set_ylabel("Error Rate")

    lists = sorted(test_error.items()) # sorted by key, return a list of tuples

    x, y = zip(*lists) # unpack a list of pairs into two tuples

    ax.set_xscale('log')
    plt.plot(x, y, color='b', label="Test Error Rate")
    ax.set_xlabel("Cost (log)")
    ax.set_ylabel("Error Rate")
    ax.set_title(f"Degree ={degree}")
    ax.legend()
    ax.grid()


plt.show()
plt.tight_layout()
```

###### There is no difference in validation error or training error fro different degrees and all the test errors for different degrees shows that the minimum value occurs at a C = 0.1.So the best values degree=2,C=0.1
