Another explanation:
We will use the alternate form of the lasso optimization problem

$$(y_1−β̂_1x_1−β̂_2x_1)^2+(y_2−β̂_1x_2−β̂_2x_2)^2 \space subject\space to\space |β̂_1|+|β̂_2|≤s.$$

Geometrically the lasso constraint take the form of a diamond centered at the origin of the plane ($β̂_1$,$β̂_2$) which intersects the axes at a distance s from the origin. 
By using the setting of this problem 
$$(x_{11}=x_{12}=x_1, x_{21}=x_{22}=x_2, x_1+x_2=0 \space and \space y_1+y_2=0), we\space have\space to\space minimize\space the\space expression$$
$$2[y_1−(β̂_1+β̂_2)x_1]^2≥0 $$
This optimization problem has a simple solution : $β̂_1+β̂_2=y_1/x_1$ 

Geometrically, this is a line parallel to the edge of the diamond of the constraints. Now, solutions to the lasso optimization problem are contours of the function $[y_1−(β̂_1+β̂_2)x_1]^2$ that intersects the diamond of the constraints. So, the entire edge $β̂_1+β̂_2=s$ (as is the edge $β̂_1+β̂_2=−s)$ is a potential solution to the lasso optimization problem. 

Thus, the lasso optimization problem has a whole set of solutions instead of a unique one :
{$(β̂_1,β̂_2):β̂_1+β̂_2=s \space with \space β̂_1,β̂_2≥0 \space and \space β̂_1+β̂_2=−s \space with \spaceβ̂_1,β̂_2≤0$}

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def loss(y1, β1, λ):
    return np.power(y1 - β1, 2) + λ*(β1**2)

y1 = 3
λ  = 2
β = np.linspace(-10.0, 10.0,50)
results = [loss(y1, β1, λ) for β1 in β]
β_solution = y1/(1+λ)

plt.scatter(x=β, y=results)
plt.axvline(x=β_solution, color='r')
plt.xlabel('β1')
plt.ylabel('Cost');
```

```python
from math import log
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import itertools
```

```python

def fit_linear_reg(X,Y):
    #Fit linear regression model and return RSS and adjusted R squared values, AIC, BIC, MAllow's Cp
    model = LinearRegression(fit_intercept = True)
    model.fit(X,Y)
    n=len(Y)
    num_params=X.shape[1]+1
    RSS = mean_squared_error(Y,model.predict(X)) * n
    adj_R_squared = 1 - (1-model.score(X, Y))*(len(Y)-1)/(len(Y)-X.shape[1]-1)
    aic = n*log(RSS)+2*num_params
    bic = n*log(RSS)+log(n)*num_params
    cp = (RSS+2*num_params*(RSS/n))/n
    return [RSS, adj_R_squared, aic, bic, cp, model]
```

```python
def get_models(X, y):
    """
    Fit all possible models that contain exactly k predictors.
    """
    RSS_list,R_squared_list,Aic_list,Bic_list,cp_list,feature_list=[],[],[],[],[],[]
    numb_features = []
    models_list=[]
    # List all available predictors
    for k in range(1,len(X.columns)+1):
        X_combos = itertools.combinations(list(X.columns), k)
        for combo in X_combos:
            tmp_result = fit_linear_reg(X[list(combo)],y)   #Store temp result 
            RSS_list.append(tmp_result[0])                  #Append lists
            R_squared_list.append(tmp_result[1])
            Aic_list.append(tmp_result[2])
            Bic_list.append(tmp_result[3])
            cp_list.append(tmp_result[4])
            models_list.append(tmp_result[5])
            feature_list.append(combo)
            numb_features.append(len(combo))   
        print('Progess: models with k = {}, done'.format(k))
    #Store in DataFrame
    df = pd.DataFrame({'numb_features': numb_features,'RSS': RSS_list, 
                   'Adj_R_squared':R_squared_list,'AIC':Aic_list,'BIC':Bic_list,
                   'Cp':cp_list,'features':feature_list,'Model':models_list})
    return df
```

```python
X_df = pd.DataFrame({'X': X**1,
                  'X2': X**2,
                  'X3': X**3,
                  'X4': X**4,
                  'X5': X**5,
                  'X6': X**6,
                  'X7': X**7,
                  'X8': X**8,
                  'X9': X**9,
                  'X10': X**10})
y_df = pd.DataFrame({'y': Y})

results_df= get_models(X_df,y_df)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def fit_linear_reg_sm(X,Y):
    #Fit linear regression model and return RSS and adjusted R squared values, AIC, BIC, MAllow's Cp
    model = sm.OLS(Y,sm.add_constant(X)).fit()
    n=len(Y)
    num_params=X.shape[1]+1
    RSS = mean_squared_error(Y,model.predict(sm.add_constant(X))) * n
    adj_R_squared = model.rsquared_adj
    aic = model.aic
    bic = model.bic
    cp = (RSS+2*num_params*(RSS/n))/n
    return [RSS, adj_R_squared, aic, bic, cp,model]
```

```python

def forward_stepwise(X, y,scorer='rss',threshold=0.01,num_features=3):
    p_used = []
    RSS_list,R_squared_list,Aic_list,Bic_list,cp_list,feature_list=[],[],[],[],[],[]
    numb_features = []
    models_list=[]
    
    
    for k in range(1,len(X.columns)+1):
        best_RSS=np.inf
        best_adj_r_squared=0
        best_cp=np.inf
        best_bic=np.inf
        best_aic=np.inf
        p_left = list(set(X.columns)-set(p_used))
        #print(p_left)
        for combo in itertools.combinations(p_left,1):
            #print(combo)
            x=X[p_used+list(combo)]
            tmp_result = fit_linear_reg_sm(x,y)
            #print(tmp_result[1])
            if scorer=='rss':
                if (tmp_result[0]< best_RSS):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
            
            elif scorer=='adj_r':
                
                if (tmp_result[1]> best_adj_r_squared): #and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_adj_r_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
                
                    
            elif scorer=='aic':
                if (tmp_result[2]< best_aic):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
                    
            elif scorer=='bic':
                if (tmp_result[3]< best_bic):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
                    
            elif scorer=='cp':
                if (tmp_result[4]< best_cp):# and (tmp_result[-1].pvalues[list(combo)].values[0]<threshold):
                    best_RSS = tmp_result[0]
                    best_R_squared = tmp_result[1]
                    best_aic = tmp_result[2]
                    best_bic = tmp_result[3]
                    best_cp = tmp_result[4]
                    best_model = tmp_result[5]
                    best_feature = combo[0]
        #print(best_feature)
        if (best_feature not in p_used) and (len(p_used)<num_features) :
            p_used.append(best_feature)
            p_left.remove(best_feature)
            RSS_list.append(best_RSS)
            R_squared_list.append(best_R_squared)
            Aic_list.append(best_aic)
            Bic_list.append(best_bic)
            cp_list.append(best_cp)
            models_list.append(best_model)
            feature_list.append(p_used.copy())
            numb_features.append(len(p_used))
        
        
    df = pd.DataFrame({'numb_features': numb_features,'RSS': RSS_list, 
                   'Adj_R_squared':R_squared_list,'AIC':Aic_list,'BIC':Bic_list,
                   'Cp':cp_list,'features':feature_list,'Model':models_list})
            

        
    return df
```

```python
def backward_stepwise(X, y, scorer='rss', results=[]):
    """Perform backward stepwise variable selection as described in
    An Introductions to Statistical Learning (p.209)
    Returns pandas dataframe object  with results for each step"""
    # List predictors that havent's been used so far
    p_all = list(X.columns)

    # Check if we're starting out here
    if not results:
        # Fit model with all features
        f     = '{} ~ {}'.format(y.columns[0], '+'.join(p_all))
        model = smf.ols(formula=f, data=pd.concat([X, y], axis=1)).fit()
        # Begin backward stepwise recursion
        if scorer == 'rss':
            return backward_stepwise(X, y, scorer, [(model.ssr, p_all)])
        if scorer == 'adj_r':
            return backward_stepwise(X, y, scorer, [(model.rsquared_adj, p_all)])
        if scorer == 'bic':
            return backward_stepwise(X, y, scorer, [(model.bic, p_all)])
        if scorer == 'aic':
            return backward_stepwise(X, y, scorer, [(model.aic, p_all)])
        if scorer == 'cp':
            RSS=model.ssr
            n= model.nobs
            cp=(RSS+2*model.df_model*(RSS/n))/n
            
            return backward_stepwise(X, y, scorer, [(cp, p_all)])
    else:
        p_used = results[-1][1]
    
    # Job done, exit recursion
    if len(p_used) == 1:
        scores = [r[0] for r in results]
        preds  = [r[1] for r in results]
        numb_features=[len(r[1]) for r in results]
        return pd.DataFrame({scorer: scores, 'predictors': preds,'numb_features':numb_features})    
    
    # Get rss score for each possible removed predictor
    r = []
    for p in p_used:
        p_test = [i for i in p_used if i != p]
        f     = '{} ~ {}'.format(y.columns[0], '+'.join(p_test))
        # Fit model
        model = smf.ols(formula=f, data=pd.concat([X, y], axis=1)).fit()
        r     += [(model, p_test)]
    
    # Choose removal of predictor which yields best score
    if scorer == 'rss':
        best_model = sorted(r, key=lambda tup: tup[0].ssr)[0]
        best_score = (best_model[0].ssr, best_model[1])
    elif scorer == 'adj_r':
        best_model = sorted(r, key=lambda tup: tup[0].rsquared_adj)[-1]
        best_score = (best_model[0].rsquared_adj, best_model[1])        
    elif scorer == 'bic':
        best_model = sorted(r, key=lambda tup: tup[0].bic)[0]
        best_score = (best_model[0].bic, best_model[1]) 
    elif scorer == 'aic':
        best_model = sorted(r, key=lambda tup: tup[0].aic)[0]
        best_score = (best_model[0].aic, best_model[1]) 
    elif scorer == 'cp':
        best_model = sorted(r, key=lambda tup: (tup[0].ssr+2*tup[0].df_model*(tup[0].ssr/tup[0].nobs))/tup[0].nobs)[0]
        best_score = ((best_model[0].ssr+2*best_model[0].df_model*(best_model[0].ssr/best_model[0].nobs))/best_model[0].nobs, best_model[1])

    new_results = results + [best_score]
    # Recursive call to self
    return backward_stepwise(X, y, scorer, new_results)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.linear_model import Ridge,Lasso
def Ridge_CV(X_train,y_train,X_test,y_test,alphas,splits):
    val_MSE={}
    test_MSE={}
    for alpha in alphas:
        mse_kf=0
        scaler = StandardScaler().fit(X_train)#https://stats.stackexchange.com/questions/86434/is-standardisation-before-lasso-really-necessary
        X_train_scaled = scaler.transform(X_train) 
        X_test_scaled = scaler.transform(X_test)
        kf = KFold(n_splits=splits)

        for train_index, val_index in kf.split(X_train_scaled):
            X_train_a, X_val_a = X_train.iloc[train_index], X_train.iloc[val_index]
            y_train_a, y_val_a = y_train.iloc[train_index], y_train.iloc[val_index]
            scaler_val = StandardScaler().fit(X_train)
            X_train_ascaled = scaler_val.transform(X_train_a)
            X_val_ascaled = scaler_val.transform(X_val_a)
            model = Ridge(alpha)
            model.fit(X_train_ascaled,y_train_a)
            mse_kf+= mean_squared_error(y_val_a,model.predict(X_val_ascaled))
        val_MSE[alpha] = mse_kf/splits
        model = Ridge(alpha)
        model.fit(X_train_scaled,y_train)
        p = model.predict(X_test_scaled)
        test_MSE[alpha] = mean_squared_error(p, y_test)
    min_val_lambda = min(val_MSE, key=val_MSE.get)
    min_test_lambda = min(test_MSE, key=test_MSE.get)
    # Plot validation MSE
    lists = sorted(val_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(121)
    plt.plot(x, y, color='r')
    plt.axvline(x=min_val_lambda,ls='--',color='b')
    plt.grid()
    ax.set_xlabel('Lambda')
    ax.set_ylabel('Validation MSE')
    ax.set_title('Validation MSE vs Lambda')
    plt.xscale('log')

    lists = sorted(test_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    ax = fig.add_subplot(122)
    plt.plot(x, y, color='g')
    plt.axvline(x=min_val_lambda,ls='--',color='r')
    plt.axvline(x=min_test_lambda,ls='--',color='b')
    plt.xscale('log')
    plt.grid()
    ax.set_xlabel('Lambda')
    ax.set_ylabel('Test MSE')
    ax.set_title('Test MSE vs Lambda')
    plt.show()
    
    return val_MSE,test_MSE

        
            
```

```python
from sklearn.preprocessing import scale 

from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression, PLSSVD
```

```python
def PCR_CV(X_train, Y_train, X_test, Y_test, M, splits):
    X_train_scaled = scale(X_train)
    X_test_scaled = scale(X_test)

    val_MSE = {}
    test_MSE = {}

    for m in M: # Iterate over number of principal components
        pca = PCA(n_components=m)
        pca = pca.fit(X_train_scaled)
        X_train_reduced = pca.transform(X_train_scaled)
        X_test_reduced = pca.transform(X_test_scaled)

        mse = 0
        kf = KFold(n_splits=splits)
        for train_index, test_index in kf.split(X_train_scaled):
            X, X_CV = X_train_scaled[train_index], X_train_scaled[test_index]
            Y, Y_CV = Y_train.iloc[train_index], Y_train.iloc[test_index]
            pca = PCA(n_components=m)
            pca = pca.fit(X_train_scaled)
            X = pca.transform(X)
            X_CV = pca.transform(X_CV)
            model = LinearRegression()
            model.fit(X, Y)
            p = model.predict(X_CV)
            mse += mean_squared_error(p, Y_CV)
        val_MSE[m] = mse/splits

        # Compute test MSE for the model
        model = LinearRegression()
        model.fit(X_train_reduced, Y_train)
        p = model.predict(X_test_reduced)
        test_MSE[m] = mean_squared_error(p, Y_test)

    min_val_components = min(val_MSE, key=val_MSE.get)
    min_test_components = min(test_MSE, key=test_MSE.get)
    # Plot validation MSE
    lists = sorted(val_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(121)
    plt.plot(x, y, color='r')
    plt.axvline(x=min_val_components,ls='--',color='b')
    plt.grid()
    ax.set_xlabel('Number of Principal Components')
    ax.set_ylabel('Validation MSE')
    ax.set_title('Validation MSE vs Principal Components')
    

    lists = sorted(test_MSE.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    ax = fig.add_subplot(122)
    plt.plot(x, y, color='g')
    plt.axvline(x=min_val_components,ls='--',color='r')
    plt.axvline(x=min_test_components,ls='--',color='b')
    
    plt.grid()
    ax.set_xlabel('Number of Principal Components')
    ax.set_ylabel('Test MSE')
    ax.set_title('Test MSE vs Principal Components')
    plt.show()
    
    return val_MSE,test_MSE
```

```python
def get_fn_value(used_cols,model,beta):
    model_coefficients=np.array(list(model.coef_[0]))
    Real_coef=np.array([beta[i] for i in used_cols])
    
    value=np.sqrt(np.sum((Real_coef-model_coefficients)**2))
    return value
    
```
